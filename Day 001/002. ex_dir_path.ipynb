#  폴더 및 파일 다루기

```python
## 모듈 로딩
import os
```

```python
# 특정 폴더 안의 모든 데이터리스트 가지고 오기
folder = '../../datas/'

#그래서 folder에 지정한 경로에 ㄹㅇ로 datas 폴더(파일)가 있냐?
#있으면 True, 없으면 False를 줘
os.path.exists(folder)
```

```python
datlist = os.listdir(folder) #그래서 이 안에 뭐 있는데? 이름을 리스트로 줘봐

for dataname in datlist:
    print(dataname, os.path.isdir(folder + dataname), sep= '---') #괄호 안의 경로에 있는 게 디렉토리(폴더)니?
```

```python
# data 폴더 아래의 파일은 파일명만, 폴더라면 폴더 안의 데이터리스트 출력해줘

for dataname in datlist:
    #이거 폴더임?
    if os.path.isdir(folder+dataname)  == True: #ㅇㅇ 맞음
        # 폴더 안의 데이터리스트 출력하기
        datalist2 = os.listdir(folder+dataname) # 그럼 폴더 안에 있는 거 리스트로 줘봐
        for data in datalist2: 
            if os.path.isdir(folder+dataname+'/'+data): 
                print('##[',data, ']is in directory')
                datalist3 = os.listdir(folder+dataname+'/'+data)
                print (datalist3)
                for data2 in datalist3: 
                    print(f'{data2}is in the folder')
        print(datalist2, 'in directory')
    else: #ㄴㄴ 아니 그냥 데이터파일이야
       print(dataname, 'is data') #그럼 데이터 이름이랑 is data를 붙여서 출력해줘
```
