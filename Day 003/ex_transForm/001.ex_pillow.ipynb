```python
## Module

from PIL import Image
import torchvision.transforms as transforms
```

```python
imgcat = '../../datas/smudgeyyy.jpg'
```

```python
cat = Image.open(fp = imgcat)
cat.format, cat.height, cat.width, cat.mode, type(cat)
```

```python
# 이미지 텐서화

totensor = transforms.ToTensor() #labelencoder처럼.... 인스턴스 만들고 집어넣기
catTs = totensor(cat)
```

```python
#변환된 이미지 텐서 확인
print(catTs.shape, catTs.ndim)
print(catTs[0])
```

```python
#### OPEN CV
import cv2
img = cv2.imread(filename=imgcat) #BGR
img2 = cv2.cvtColor(src=img, code=cv2.COLOR_BGR2RGB) #RGB
type(img)

#tensorize
cat2 = totensor(img)
cat3 = totensor(img2)
print(cat2.shape, cat2.ndim, cat2[0], sep='\n\n')
print(cat3.shape, cat3.ndim, cat3[0], sep='\n\n')
```

```python
### resizing
preprocessing = transforms.Compose(
    [transforms.Resize(size=(50,50)), # 넘파이 안됨..
    transforms.ToTensor()]
    )
```

```python
preprocessing(cat)
```

### 이미지 데이터셋 <hr>
- torchvision.ImageFolder 클래스 사용
    - 이미지 데이터 라벨링
    - 이미지 데이터 전처리
        

```python
from torchvision.datasets import ImageFolder
```

```python
imgroot = '../../../datas/Project Jellyfish'
imgdataset = ImageFolder(root=imgroot
                  , transform=preprocessing)
```

```python
# 폴더명이 분류 클래스
imgdataset.classes, imgdataset.class_to_idx, imgdataset.imgs
```

```python
# 데이터 로드
from torch.utils.data import DataLoader
imgdl = DataLoader(dataset= imgdataset)
```

```python
for (img, label) in imgdl:
    print(img.shape, label.shape)
    break
```
